```python
import numpy as np
import matplotlib.pyplot as plt

# Mandelbrot set function
def mandelbrot(threshold=50, x_min=-2, x_max=1, y_min=-1.5, y_max=1.5, resolution=500):
    # Create a grid of complex numbers
    x, y = np.mgrid[x_min:x_max:resolution*1j, y_min:y_max:resolution*1j]
    c = x + 1j * y
    z = np.zeros_like(c)
    mask = np.ones(c.shape, dtype=bool)
    
    for i in range(threshold):
        z[mask] = z[mask]**2 + c[mask]
        mask[np.abs(z) > threshold] = False
    
    plt.imshow(mask.T, extent=[x_min, x_max, y_min, y_max], cmap="gray")
    plt.savefig('mandelbrot.png')
    plt.show()

# Generate Mandelbrot set image
mandelbrot()

import numpy as np

# Create a random transition matrix and normalize each row
def create_transition_matrix(size=5):
    P = np.random.rand(size, size)
    P /= P.sum(axis=1, keepdims=True)
    return P

# Create a random probability vector and normalize it
def create_probability_vector(size=5):
    p = np.random.rand(size)
    p /= p.sum()
    return p

# Main function to compute p after 50 transitions and the stationary distribution
def markov_chain():
    P = create_transition_matrix()
    p = create_probability_vector()
    
    # Apply the transition rule 50 times
    for _ in range(50):
        p = np.dot(P.T, p)
    
    # Find stationary distribution
    eigvals, eigvecs = np.linalg.eig(P.T)
    stationary = eigvecs[:, np.isclose(eigvals, 1)]
    stationary = stationary / stationary.sum()
    
    # Check if p50 matches the stationary distribution
    difference = np.abs(p - stationary.flatten())
    print("Difference:", difference)
    print("Match within 10^-5:", np.all(difference < 1e-5))

# Run the Markov chain process
markov_chain()
import numpy as np
import sympy as sp
import pandas as pd
import time
import matplotlib.pyplot as plt

# Define the Taylor series approximation function
def taylor_approx(func, start, end, degree, fixed_c):
    x = sp.symbols('x')
    # Compute Taylor series up to specified degree
    series = sum(func.diff(x, n).subs(x, fixed_c) / sp.factorial(n) * (x - fixed_c)**n for n in range(degree + 1))
    # Convert the series to a numpy-compatible function
    f_approx = sp.lambdify(x, series, 'numpy')
    return f_approx(np.linspace(start, end, 100))

# Run the Taylor series approximation with specific parameters and plot the results
def run_taylor_series():
    x = sp.symbols('x')
    func = x * sp.sin(x)**2 + sp.cos(x)  # Define the function f(x)
    approximation = taylor_approx(func, -10, 10, 99, 0)  # Compute the Taylor approximation

    # Plot the original function and approximation
    x_vals = np.linspace(-10, 10, 100)
    f_exact = sp.lambdify(x, func, 'numpy')(x_vals)
    plt.plot(x_vals, f_exact, label='Exact Function')
    plt.plot(x_vals, approximation, label='Taylor Approximation', linestyle='--')
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()

# Advanced function for varying degrees and saving results to CSV
def advanced_taylor_approx():
    x = sp.symbols('x')
    func = x * sp.sin(x)**2 + sp.cos(x)  # Define the function f(x)
    f_exact = sp.lambdify(x, func, 'numpy')(np.linspace(-10, 10, 100))
    rows = []  # List to store each row of results

    # Iterate through degrees and calculate the approximation, error, and computation time
    for degree in range(50, 101, 10):  # Degrees from 50 to 100 in steps of 10
        start_time = time.time()
        approximation = taylor_approx(func, -10, 10, degree, 0)  # Compute approximation
        time_elapsed = time.time() - start_time  # Measure computation time
        error = np.sum(np.abs(f_exact - approximation))  # Calculate error
        
        # Append the results for this degree to the rows list
        rows.append({"Degree": degree, "Error": error, "Time (s)": time_elapsed})

    # Convert the list of results to a DataFrame
    df = pd.DataFrame(rows)

    # Write the DataFrame to a CSV file
    df.to_csv("taylor_values.csv", index=False)  # Save results to taylor_values.csv

# Run the Taylor series functions
run_taylor_series()
advanced_taylor_approx()
```
